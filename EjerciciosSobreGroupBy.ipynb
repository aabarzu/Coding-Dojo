```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd

filename = '/content/drive/MyDrive/CodingDojo/DataScience/PandasForDataManipulation/Datos/CalIt2.data'     # csv

df = pd.read_csv(filename ,header=None)   # csv

df.head(10) 

```

## **Cambiar el nombre de las columnas**

```python
#Cambiar el nombre de las columnas
#df = df.rename(columns={'0':'ID_Flujo'}) # Prerguntar por que no funciona

df.columns = ['ID_Flujo','Fecha','Hora','Conteo']

df['Fecha'] = pd.to_datetime(df['Fecha']) # Convierto a fecha 

df
```

## **Seleccionar datos**

```python
#Seleccionar datos

#1.-Selecciona todos los datos para la fecha del 24 de Julio de 2005 que tengan un número de identificación de flujo = 7

filter_fecha = df['Fecha'] =='2005-07-24'

#De los datos seleccionados en el paso 1, clasifica los resultados de conteo en orden descendiente y
#Crea el formato de salida de las 10 filas superiores. 
#Asigna el formato de salida a una variable llamada top10.

top10 = df.loc[filter_fecha].sort_values('Conteo',ascending=False).head(10)

top10
```

## **Función Apply**

```python
#1.-Para las 10 filas sacadas arriba, usa la función Apply de pandas para restar el valor menor de los 10 de todos ellos y luego saca el valor promedio de los conteos resultantes.

top10['Conteo'] = top10['Conteo'].apply(lambda x:x-top10['Conteo'].min())

top10


```

```python
#luego saca el valor promedio de los conteos resultantes.
print(f" Suma   = {top10['Conteo'].sum()}  ")
print(f" Conteo = {top10['Conteo'].count()}")
print(f" El promedio de los conteos resultantes es {top10['Conteo'].sum()/top10['Conteo'].count()}")
```

## **Agrupar**

```python
#1.-Selecciona los datos del mes de agosto 2005 que tengan una ID de flujo = 7

fillter_Data = (df['Fecha']  >= '2005-08-01') & (df['Fecha']  < '2005-09-01') & (df['ID_Flujo'] == 7)

#2.-Agrupa los datos basándose en la fecha y saca el conteo máximo por fecha.

df.loc[fillter_Data].groupby(['Fecha'])[['Conteo']].max('Conteo')
```
