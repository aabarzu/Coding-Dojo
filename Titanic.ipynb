```python
#Mostrar por pantalla los nombres de las personas que iban en primera clase ordenadas alfabéticamente.

titanic_filter = titanic_df['Pclass'] == 1

print(titanic_df[titanic_filter].sort_values('Name')['Name'])

print(titanic_df.loc[titanic_filter , :].sort_values('Name')['Name'])

print(titanic_df.loc[titanic_df['Pclass'] == 1].sort_values('Name')['Name'])
```

```python
#Mostrar por pantalla el porcentaje de personas que sobrevivieron y murieron.

titanic_df.groupby(['Survived']).count()['PassengerId']/titanic_df['Survived'].count()*100.0 

```

```python
#Mostrar por pantalla el porcentaje de personas que sobrevivieron en cada clase.
titanic_df.groupby(['Pclass']).count()['Survived']/titanic_df['Survived'].count()*100.0 




```

```python
#Eliminar del DataFrame los pasajeros con edad desconocida.
data_None=titanic_df['Age'].isna()
titanic_df=titanic_df.loc[~data_None]
titanic_df
```

```python
#Mostrar por pantalla la edad media de las mujeres que viajaban en cada clase.
titanic_df.groupby(['Sex']).mean('Age')['Age']
```

```python
#Añadir una nueva columna booleana para ver si el pasajero era menor de edad o no
#titanic_df=titanic_df.drop(columns=['Younger'])

titanic_df.insert(12,'Younger', True )

titanic_df.insert(12,'Younger', True )
titanic_df.loc[titanic_df['Age'] >= 18 , 'Younger'] = False
titanic_df

```
