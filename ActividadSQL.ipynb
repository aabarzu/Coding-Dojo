```python
# imports
import pandas as pd
from sqlalchemy import create_engine

```

```python
# cargar datos: ¡cambia la ruta para que coincida con tu unidad!
school = pd.read_csv('/content/drive/CHANGETHEPATH/School Details.csv')
# Asegúrate de que estén cargados correctamente
school.head(2)
```

```python
# crear motor sqlite
engine = create_engine('sqlite:///school.db', echo=True)
# crear conexión con el motor
conn = engine.connect()
```

```python
# agregar marco de datos como una tabla en sqlite
sqlite_table = "school_details"
school.to_sql(sqlite_table, conn, if_exists='fail')
```

```python
def EjecutaSQSL(sql,filas = 5):
  # Escribe tu consulta SQL en esta cadena

  # Lee la consulta sql en pandas para ver los resultados
  df = pd.read_sql_query(sql, engine)
  return df.head(filas)
```

```python
def EjecutaSQSL_(sql):
  # Escribe tu consulta SQL en esta cadena

  # Lee la consulta sql en pandas para ver los resultados
  conn.execute(sql)
  
```

```python
# Escribe tu consulta SQL en esta cadena
sql = """
select * from school_details
"""
EjecutaSQSL(sql) 
```

**2.-Muestra el NombreEstablecimiento y DistanciaDistrito (en km) de colegios en los que DistanciaDistrito (en km) es más de 100 km.0**

```python
# Escribe tu consulta SQL en esta cadena
sql = """
select EstablishmentName,[DistrictHQDistance (in Km)]  from school_details where [DistrictHQDistance (in Km)] >100
"""
EjecutaSQSL(sql,100) 
```

**3.-Muestra NombreEstablecimiento, NivelEscolar, DistanciaDistrito (en km) de colegios donde el nivel escolar es Junior High School.**

```python
# Escribe tu consulta SQL en esta cadena
sql = """
select EstablishmentName,SchoolLevel,[DistrictHQDistance (in Km)] from school_details where SchoolLevel = 'JHS'
"""
EjecutaSQSL(sql,100) 
```

**4.- Muestra NombreEstablecimiento, NivelEscolar, DistanciaDistrito (en km) de colegios que contengan las palabras “Junior High” en su nombre.**

```python
# Escribe tu consulta SQL en esta cadena
sql = """
select EstablishmentName,SchoolLevel,[DistrictHQDistance (in Km)] from school_details where EstablishmentName like '%Junior High%'
"""
EjecutaSQSL(sql,100) 
```

**5.- Muestra NombreEstablecimiento, NivelEscolar, DistanciaDistrito (en km) de colegios cuyo nombre comience con la letra “C”. Ordena los resultados de forma alfabética por nombre de colegio  y limita los resultados de la consulta a 10 líneas.**

```python
# Escribe tu consulta SQL en esta cadena
sql = """
select EstablishmentName,SchoolLevel,[DistrictHQDistance (in Km)] from school_details where EstablishmentName like 'C%' order by EstablishmentName
LIMIT 10 
"""
EjecutaSQSL(sql,100) 
```
