```python
from google.colab import drive
drive.mount('/content/drive') #  force_remount=True , pide eso no sé porqué
import pandas as pd

filename = '/content/drive/MyDrive/CodingDojo/DataScience/PandasForDataManipulation/Datos/catsvdogs.xlsx' # Excel

df = pd.read_excel(filename) # Excel

df.head() 
```

```python
# Librerias para graficar
import matplotlib.pyplot as plt
```

```python
df.info()
```

```python
df.describe()
```

```python
data_dog = df['Percentage of Dog Owners']
data_cat = df['Percentage of Cat Owners']

fig1 = plt.figure(dpi=100)
ax1  = fig1.add_subplot(1,1,1)
ax2  = fig1.add_subplot(1,1,1)
ax3  = fig1.add_subplot(1,1,1)
ax1.hist(data_dog.values , edgecolor='black' ,linewidth = 0.5 )
ax2.hist(data_cat.values , edgecolor='black', linewidth = 0.5 ,alpha = 0.8)
plt.legend(labels=['Dogs', 'Cats'])

plt.ylabel("Count")
plt.xlabel("Percentage of Animal Owners")
#plt.style.use('classic')
plt.grid();
plt.show();
```

```python
data_dog = df['Mean Number of Dogs per household'].sort_values()
data_cat = df['Mean Number of Cats'].sort_values()

datos = [  data_cat.values , data_dog.values ]

fig = plt.figure(dpi=100) 
ax = fig.add_subplot(111) 

medianprops = dict( linewidth=4, color='#ffed4d')

bp = ax.boxplot(datos, patch_artist = True, notch ='True',labels=['Cats','Dogs'] , medianprops = medianprops , widths = 0.8) 

colors = ['#f2ce90', '#8b8cf5'] 
          
for patch, color in zip(bp['boxes'], colors): 
    patch.set_facecolor(color) 

plt.ylabel('Mean Number Of Animals per State')
plt.show(bp) 
```

```python
data_dog = df['Mean Number of Dogs per household'].sort_values()
data_cat = df['Mean Number of Cats'].sort_values()

datos = [  data_cat.values , data_dog.values ]
plt.style.use('seaborn')

boxprops = dict(color='blue' , facecolor='#8b8cf5')
medianprops = dict(linestyle='-', linewidth=4, color='y')

plt.figure(dpi=100)
plt.boxplot( datos , notch=True , labels=['Cats','Dogs'] ,patch_artist=True, medianprops=medianprops ,  boxprops=boxprops ,widths = 0.8)
plt.ylabel('Mean Number Of Animals per State')

plt.show()
```
